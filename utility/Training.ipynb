## <span style="color:orange"> Welcome to PRIvacy preserving data SIMulator (PRISIM) Toolkit - Part 1 </span>

This toolkit provides an end-to-end solution for generating a synthetic version of a real-world sensitive dataset and running a privacy engine on top of it to make sure that the data is private, i.e., no private and sensitive information is leaked that compromises the re-identification of the constituting subjects. It does so while preserving its underlying covariate structure so that it is still useful for downstream analytics tasks.

The end-to-end pipeline is split into 2 phases to de-couple the longer training process from the faster generation process. Details of each phase are presented below :
1. <span style="color:indigo"> **Training Pipeline:** </span> this phase takes care of obtaining the original data, making sure the correct data types are specified for each of the features, and performing some EDA to get a sense of the data landscape and patterns, and finally train a Generative model on the data. The Generative Model is then saved to be used in the second part of the pipeline to generate synthetic private data.


2. <span style="color:indigo"> **Private Synthetic Data Generation Pipeline:** </span> this part reads the saved model from the previous phase to generate synthetic private samples on the fly. This phase also provides a comprehensive evaluation suite for evaluating the quality of the private data.


The detailed steps of each pipeline are outlined below: 
1. <span style="color:indigo"> **Training pipeline :** </span>
        1.1. Uploading the original data
        1.2. Verifying the features' data types
        1.3. Exploring the data properties (Exploratory Data Analysis)
        1.4. Training the Generative Model on the original data


2. <span style="color:indigo"> **Private Synthetic Data Generation pipeline:** </span>
        2.1. Loading the saved generative model from the Training phase.
        2.2. Private Data Generation
        2.3. EDA on the Private Data
</details>
