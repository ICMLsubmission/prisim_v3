### <span style="color:indigo"> Step 2.2.2 : Generate private data </span>

<br/>

<details>
<summary><b> <<  CLICK HERE to expand / collapse Section Overview >> </b></summary>


In this step we generate private synthetic data with different privacy levels. It is important to note that there is a trade-off between privacy and ML utility (higher privacy usually leads to lower ML performance).

***Privacy*** is measured here in terms of the proportion of the samples that have higher distance from their closest original counterpart, therefore making them harder to re-identify. For instance, if a synthetic dataset has high number of samples that are very close to the original data, then it stands at a higher risk of data leakage thus resulting low privacy and vice versa. 

We run different privacy scenarios and report the corresponding privacy and ML utility for each scenario. Based on the requirement, we can choose the suitable scenario to generate the final private synthetic data. 

**Note:** The %ML utility shown here is not the actual ML utility but rather *ML utility relative to the original data utility*. This implies that if ML utility for the task when trained on original data was 75% and on synthetic data is 60% then the reported number is 80% (60/75). The graph essentially tells  ***how much of the original ML utility can be retained for various privacy scenarios ?***.

In this step we use the private data from the chosen scenarion above to dive deeper into its ML utility and generation quality. 

**Ml utility:** We compare the ML performance of real data with synthetic private data. We do so by keeping a common held-out test set from the original data and training on the real data (train set) as well as on the private data. 

**Generation Quality Index (GQI):** We also introduce an objective metric called GQI to quantify how well the private data is able to retain the statistical properties of the original data. GQI measures the similarity of the distributions of bi-variate correlations between original and private data. The higher, the better, i.e, for instance 98% GQI implies that the generated synthetic data preserves 98% correlations of the original data.

**Feature importance:** although similar overall ML task performance between real and synthetic data is a good check of the utility, it is not sufficient. It is also important that the realtive strength of features that contribute towards the predictive performance remains intact. 

To that end, this step also plots the relative feature importance of the models trained on real and synthetic data. Ideally, the relative order of top important features should be the same, although minor 1-position switches is acceptable.

To run this section:

2. This triggers the 2 ML models in the background - one trained on real data, and other on synthetic private data.
3. The ML utility scores of the 2 models is displayed, along with the GQI and privacy level of the chosen scenario. 
4. Finally, the **feature importance** curve is plotted to compare the relative order of important features of the 2 datasets.
